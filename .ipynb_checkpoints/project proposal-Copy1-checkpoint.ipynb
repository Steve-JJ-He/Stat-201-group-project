# Does Temperature of Cities Within the Same Latitude Differs?

## Introduction

As climate change worsens, climate factors become a  consideration for immigrants to choose which city or country for their relocation. Climigration is a phenomenon that refers to forced permanent migration due to climate change. A report from the World Bank has stated that "216 million people will have to migrate within their own countries by 2050" (Clement et al.). Factors such as temperature, humidity, and windspeed are the primary consideration. 

Latitude is a geographic coordinate that indicates the north-south position of the equator, where the equator is 0 degrees in latitude. The closer the latitude is to the equator (lower latitude), the warmer the temperatures are (Lu Aigang, Wang Tianming, Kang Shichang, & Pang Deqian, 2009). Since latitude is one of the predominant factors that affect temperature, our null hypothesis suggests that places with the same latitude should have the same temperature. Hence, **we aim to seek whether cities of the same latitude differ in temperature.**  We hope that the findings of this research could give future migrants more information on how to choose the geographical locations when condsidering migrating within country.

This project analyzes weather data, taken from Historical Weather Data from 2012-2017, in three different cities in the United States: San Francisco, Denver, and New York. Three locations are chosen as they have similar latitude but different longtitude. 

# Preliminary Result

```r
library(tidyverse)
library(datateachr)
library(repr)
library(digest)
library(infer)
library(grid)
library(gridExtra)
library(ggplot2)
library(knitr)
library(infer)
library(inferr)
```

```r
#load original data

temperature <- read.csv("temperature.csv")
head(temperature)
```

```r
#Data Clean up
#select data only from San Fransisco, Denver and New York
#filter out data that are not available
#change the temperature unit from Kelvin to Celcius
temp <- temperature %>%
select(San.Francisco, Denver, New.York)%>%
filter(San.Francisco !="NA" , Denver !="NA" , New.York !="NA")%>%
mutate(San.Francisco = San.Francisco - 273.15, Denver= Denver - 273.15, New.York = New.York - 273.15)
head(temp)
```

```r
# mutate data frame such that there is only two columns. 
# First column is the temperature
# Second column is the city
cities <- c("SF", "Denver", "NY")
SF_temp <- temp %>% select(San.Francisco) %>% rename(temperature = San.Francisco) %>%mutate(city = cities[1]) 
Den_temp <- temp %>% select(Denver) %>% rename(temperature = Denver) %>%mutate(city = cities[2]) 
NY_temp <- temp %>% select(New.York) %>% rename(temperature = New.York) %>%mutate(city = cities[3]) 
temp_clean <- rbind(SF_temp, Den_temp, NY_temp) %>% group_by(city)
head(temp_clean)
```

```r
#draw boxplot
ggplot(temp_clean, aes(x=city, y=temperature, fill=city)) + 
   ggtitle("Boxplot of Temperature (in celcius) of Each City") +
   geom_boxplot(alpha=0.5) +
   theme_classic() 
```

```r
# Mean and Standard Deviation of temperature in each city

all_mean <- temp_clean %>% summarise(mean=mean(temperature), std=sd(temperature), Count=n())
all_mean
```

## Bootstrapping Method
We use bootrapping to construct 90% CI of temperature mean for each city

```r
#find bootstrapped temperature mean, sd 
set.seed(1) #for reproducibility
sf_bootstrap <- temp_clean %>% 
                filter(city == "SF") %>% 
                rep_sample_n(size=400, reps=3000, replace= TRUE) %>%
                group_by(replicate)%>%
                summarize(mean_sf=mean(temperature), 
                            sd_sf=sd(temperature))

den_bootstrap <- temp_clean %>% 
                filter(city == "Denver") %>% 
                rep_sample_n(size=400, reps=3000, replace= TRUE) %>%
                group_by(replicate)%>%
                summarize(mean_den=mean(temperature), 
                            sd_den=sd(temperature))

ny_bootstrap <- temp_clean %>% 
                filter(city == "NY") %>% 
                rep_sample_n(size=400, reps=3000, replace= TRUE) %>%
                group_by(replicate)%>%
                summarize(mean_ny=mean(temperature), 
                            sd_ny=sd(temperature))

head(cbind(sf_bootstrap,den_bootstrap[2], den_bootstrap[3] , ny_bootstrap[2], ny_bootstrap[3]))
```

```r
#using bootstrap
#90% CI for every city's temperature mean
ci_sf <- sf_bootstrap %>% 
    summarize(ci_lower = quantile(mean_sf, 0.05),
              ci_upper = quantile(mean_sf, 0.95))

ci_den <- den_bootstrap %>% 
    summarize(ci_lower = quantile(mean_den, 0.05),
              ci_upper = quantile(mean_den, 0.95))

ci_ny <- ny_bootstrap %>% 
    summarize(ci_lower = quantile(mean_ny, 0.05),
              ci_upper = quantile(mean_ny, 0.95))

#combine into one table
ci_boot<- cbind(cities, rbind(ci_sf, ci_den, ci_ny))
ci_boot
```

```r
#obeserved test mean from sampling distribution
# use for x intercept in ci plot
obs_mean_sf <-mean(sf_bootstrap$mean_sf)
obs_mean_den <-mean(den_bootstrap$mean_den)
obs_mean_ny <-mean(ny_bootstrap$mean_ny)

observer_test_stat <- c(obs_mean_sf, obs_mean_den, obs_mean_ny)
cbind(cities, observer_test_stat)
```

```r
#visualize bootstrap distirbution with 90% CI

ci_plot_sf <-  sf_bootstrap %>% 
    ggplot(aes(x = mean_sf)) +
    geom_histogram( colour = "white", fill = "grey") +
    annotate("rect", xmin = ci_sf$ci_lower, xmax = ci_sf$ci_upper, ymin = 0, ymax = Inf,
             fill = "deepskyblue",
             alpha = 0.3) +
    geom_vline(xintercept = obs_mean_sf,
               size = 2,
               colour = "red") +
    labs(title = "Bootstrap distribution with 90% CI temperature mean in San Francisco",
         x = "average temperature(celsius)") +
    theme_bw() 


ci_plot_den <-  den_bootstrap %>% 
    ggplot(aes(x = mean_den)) +
    geom_histogram( colour = "white", fill = "grey") +
    annotate("rect", xmin = ci_den$ci_lower, xmax = ci_den$ci_upper, ymin = 0, ymax = Inf,
             fill = "deepskyblue",
             alpha = 0.3) +
    geom_vline(xintercept = obs_mean_den,
               size = 2,
               colour = "red") +
    labs(title = "Bootstrap distribution with 90% CI temperature mean in Denver",
         x = "average temperature(celsius)") +
    theme_bw() 


ci_plot_ny <-  ny_bootstrap %>% 
    ggplot(aes(x = mean_ny)) +
    geom_histogram( colour = "white", fill = "grey") +
    annotate("rect", xmin = ci_ny$ci_lower, xmax = ci_ny$ci_upper, ymin = 0, ymax = Inf,
             fill = "deepskyblue",
             alpha = 0.3) +
    geom_vline(xintercept = obs_mean_ny,
               size = 2,
               colour = "red") +
    labs(title = "Bootstrap distribution with 90% CI temperature mean in New York",
         x = "average temperature(celsius)") +
    theme_bw() 

 
ci_plot_sf
ci_plot_den
ci_plot_sf
```

```r
#boxplot
#rename columns so that they have the same column name, 
#because we are doing rbind, so it must have the same column name
bp_sf <-rename(sf_bootstrap, bootstrap_mean = mean_sf, bootstrap_sd = sd_sf) %>% mutate(city = cities[1]) 
bp_den <-rename(den_bootstrap, bootstrap_mean = mean_den, bootstrap_sd = sd_den) %>% mutate(city = cities[2]) 
bp_ny <- rename(ny_bootstrap, bootstrap_mean = mean_ny, bootstrap_sd = sd_ny) %>% mutate(city = cities[3]) 
bp_data <- rbind(bp_sf, bp_den, bp_ny)
head(bp_data)

#draw boxplot
ggplot(bp_data, aes(x=city, y=bootstrap_mean, fill=city)) + 
   ggtitle("Boxplot of bootstrap distribution mean") +
   geom_boxplot(alpha=0.3) +
   theme_classic() 
#boxplot( bp_data$samp_dist_mean~bp_data$city)
```

## Asymptotic Method
Assume that CLT is applicable to original sample (temp_clean) 

```r
#refer to code cell 6. 
#summary of temp_clean
all_mean

#make into new variables
sf_mean <- all_mean$mean[3]
sf_sd <- all_mean$std[3]
sf_count <- all_mean$Count[3]

den_mean <- all_mean$mean[1]
den_sd <- all_mean$std[1]
den_count <- all_mean$Count[1]

ny_mean <- all_mean$mean[2]
ny_sd <- all_mean$std[2]
ny_count <- all_mean$Count[2]
```

```r
#using CLT
#find ci 90% for mean temperature in each city
sf_mean_temperature_ci <- 
tibble(lower_ci = qnorm(0.025, sf_mean, sf_sd / sqrt(sf_count)),
       upper_ci = qnorm(0.95, sf_mean, sf_sd / sqrt(sf_count)))

den_mean_temperature_ci <- 
tibble(lower_ci = qnorm(0.025, den_mean, den_sd / sqrt(den_count)),
       upper_ci = qnorm(0.95, den_mean, den_sd / sqrt(den_count)))

ny_mean_temperature_ci <- 
tibble(lower_ci = qnorm(0.025, ny_mean, ny_sd / sqrt(ny_count)),
       upper_ci = qnorm(0.95, ny_mean, ny_sd / sqrt(ny_count)))

ci_clt<- cbind(cities, rbind(sf_mean_temperature_ci, den_mean_temperature_ci, ny_mean_temperature_ci))
```

## Hypothesis Testing

#### $H_0: \mu = \mu_{s} = \mu_{d} = \mu_{n}$ vs $H_A:$ at least one of the city have different temperature mean <br> <br>
Since our project will involves 3 different cities, we are going to ANOVA to see if there is a statistical difference between their mean temperatures. <br> Since we our data have unequal variances, we cannot use One-Way ANOVA as it will violate some assumptions. <br> Hence, we will use Welch's ANOVA.


```r
oneway.test(temperature ~ city, data = temp_clean, var.equal = FALSE)
```

#### Interpretation of Welch 
##### Welch F statistic = $\frac{\text{variance between groups}}{\text{variance within groups}}$
If F statistics is close to 1, the mean of different groups are similar. <br> If it is far from 1, the mean of different groups are more different
<br> In our case F = 4809.8. Hence the mean temperature between cities are different.
##### P-value
Our p value verifies that the difference is significant. <br>
Since our $p=  2.2^{-16} < 0.001$, 
#### it tells us that there is very strong evidence against the null hyppotesis and in favor of the alternative

# Methods


To determine how geographical factors affect the temperature in North America, three cities from different regions (western, central, and eastern United States) provide variation in longitude. We bootstrapped the data to get a suitable estimate of the mean temperature for each city. Comparing the confidence interval for bootstrapping and the Asymptotic method shows that the confidence interval for bootstrapping is wider compared to the CI calculated using CLT.
 
**The findings from the data suggests that the temperature between the cities will be different, despite the cities' location in similar latitudes.** One limitation we have is that one-way ANOVA can only determine if two or more groups differ from each other. But it doesn't specify which groups are different from one another. We can add a test for the normality and homogeneity of variance to see if the distributions are comparable and better test our hypothesis. Also, we should add Post Hoc Test using the Games-Howell multiple comparison methods to know which between group differencences are statistically significant. Next, we can also analyze the temperature difference between the three selected cities in summer and winter separately so that the variance and uncertainty can be further decreased. 
 
The findings from the data will have an impact on the understanding of temperature differences in geographical regions with similar latitudes. Future research can be done with the inclusion of differing longitude in mind, as well as investigating other key geographical feature that causes the temperature difference.
